```python
import json

with open('/content/books.json', 'r') as f:
    books_data = json.load(f)

display(books_data)
```

```python
for book in books_data['books']:
    print(book['title'])
```

```python
for book in books_data['books']:
    print(f"Title: {book['title']}, Available Copies: {book['details']['available']}")
```

```python
total_copies = 0
for book in books_data['books']:
    total_copies += book['details']['copies']

print(f"Total number of copies for all books: {total_copies}")
```
